# Combining Data

```python
# whenever we create a new dataset like this that we want to be independent from our original dataset we need to run copy()
df1 = df.query('Airline == "American Airlines Inc."').copy()
df2 = df.query('Airline == "Envoy Air"').copy()

df_stack = pd.concat([df1, df2])

df1 = df1.reset_index(drop=True)
df2 = df2.reset_index(drop=True)

df_side = pd.concat([df1, df2], axis = 1)
```

# Merge Data

```python
df1 = df.groupby(['Airline', 'FlightDate'])[['DepDelay']].mean().reset_index()
df2 = df.groupby(['Airline', 'FlightDate'])[['ArrDelay']].mean().reset_index()
```

```python
# this will take any similar columns between the 2 datasets and combine on those
df1.merge(df2, how='inner')
```

```python
# if there are columns with the same name on both df, the system will automatically rename those columns
# by adding _x or _y suffixes
pd.merge(df1, df2, on = ['Airline'])
```

```python
# we can change the suffixes as we want using
pd.merge(df1, df2, on = ['Airline'], suffixes = ('_dep', '_arr'))
```

```python
# if the dataframes are different we can use left_on and right_on
pd.merge(df1, df2, left_on = ['Airline'], right_on = ['Airline'])
```
